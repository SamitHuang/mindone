```python
# integrate webdataset into mindspore dataset

class MyIterable:
    def __init__(self, urls, num_samples=-1, repeat=True, shuffle=True):
        # if shuffle, shuffle in each epoch
        if shuffle:
            self.ds = (
                wds.WebDataset(urls) # repeat = True?
                .shuffle(1000)
                .decode("rgb")
                .to_tuple("jpg;png", "json")
            )
        else:
            self.ds = (
                wds.WebDataset(urls) # repeat = True?
                .decode("rgb")
                .to_tuple("jpg;png", "json")
            )
        #self.webdatasets = webd
        self.length = num_samples
        self.ds.length = num_samples
        
    def __next__(self):
        image, data = next(iter(self.ds))
        return (image, data['caption']) # TODO: the key name can be caption or text
    
    def __iter__(self):
        
        ''' Use it for multiple sources (some online, some offline)
        sources = [iter(ds) for ds in self.datasets]
        while True:
            for source in sources:
                try:
                    yield next(source)
                except StopIteration:
                    return
        '''
        #return self.ds.iterator()
        
        for image, data in self.ds:
            try:
                yield (image, data['caption'])
            except StopIteration:
                return
            
    def __len__(self):
        return self.length


import mindspore as ms
#my_url = "/Volumes/LIN RUI/LAION/part2_small.tar"
my_url = "https://huggingface.co/datasets/jasonhuang23/sd2.1_base_train/resolve/main/part_1/00001.tar"
ms_dataset = ms.dataset.GeneratorDataset(source=MyIterable(
                                        urls=my_url, 
                                        num_samples=9), 
                              column_names=["image", "text"])

diter = ms_dataset.create_tuple_iterator(num_epochs=2, do_copy=False)

read_num = 10
for ei in range(2):
    for i, item in enumerate(diter):
        print(i, item[0].sum(), item[1])
        if i >= read_num:
            break
    print("")
  
```
